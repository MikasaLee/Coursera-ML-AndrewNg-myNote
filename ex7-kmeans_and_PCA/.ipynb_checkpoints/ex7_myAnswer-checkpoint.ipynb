## 1.1  Implementing K-means

### 1.1.1 Finding closest centroids

```python
def init2DCentroids(data,k):  # 取data前k个
    return data[:k].copy()
def randomInit2DCentroids(data,k):
    m = data.shape[0]
    return data[np.random.permutation(range(m))[:k],:]

# 我并没有去算K-means的代价，所以不需要开根号，如果要计算代价的话是要开根号的。
# def getDistance(x,u):
#     return math.sqrt(sum((x-u)**2))

def getDistance(x,u):
    '''
    x.shape:(m,k)
    u.shape:(k)
    '''
    return np.sum((x-u)**2,axis=1)

def findClosestCentroids(data,centroids):
    m = data.shape[0]
    k = centroids.shape[0]
    closest = np.zeros(m,dtype=int)
    minDis = None
    
    # 同样
#     for i in range(m):
#         x = data[i]
#         u = centroids[0]
#         minDis = getDistance(x,u)
#         res = 0
#         for j in range(1,k):
#             u = centroids[j]
#             curDis = getDistance(x,u)
#             if minDis > curDis:
#                 minDis = curDis
#                 res = j
#         closest[i] = res
    for i in range(k):
        if minDis is None:
            minDis = getDistance(data,centroids[i])
            closest[:] = i
        else:
            distances = getDistance(data,centroids[i])
            changed = minDis > distances
            closest[changed] = i
            minDis[changed] = distances[changed]
    return closest
```

### 1.1.2  Computing centroid means

```python
def computeCentroids(data,closest,k):
    m,n = data.shape
    new_centroids = np.zeros(shape=(k,n))   # 每一行为一个中心，第一列表示所有x0的和，第二列表示所有x1的和，第三列表示离该中心最近的个数
    numbers = np.zeros(shape=k,dtype=int)
    # 同样先分类后处理
#     for i in range(m):
#         new_centroids[closest[i]] += data[i]
#         numbers[closest[i]] +=1
#     for i in range(k):
#         if numbers[i] !=0:
#             new_centroids[i] /= numbers[i]
    for i in range(k):
        new_centroids[i] = np.sum(data[closest == i],axis=0)
        numbers[i] = np.sum(closest == i)
        if numbers[i] !=0:
            new_centroids[i] /= numbers[i]
    return new_centroids,numbers
```

## 1.2 K-means on example dataset

```python
def runKmeans(data,k,times,centroids = None,isRandom=False):
    if centroids is None:
        if isRandom is False: centroids = init2DCentroids(data,k)
        else: centroids = randomInit2DCentroids(data,k)
    for i in range(times):
        closest = findClosestCentroids(data,centroids)
        old_centroids = centroids
        centroids,numbers = computeCentroids(data,closest,k)
    return closest,centroids,old_centroids,numbers    # old_centroids 和 closest 是同步的，而 centroids 是根据 closest 移动更新的
closest,centroids,old_centroids,numbers = runKmeans(x,3,6)
closest,centroids,old_centroids,numbers
```

```python
draw_Kmeans(x,closest=closest,centroids=old_centroids,title='Figure 1: The expected output')
```

作业中要求Figure 1还要标出每次中心移动的轨迹，我这里是通过画多个图，每个图代表每一次迭代来实现了。

```python
def showFigure1(data,k,times):
    centroids = None
    for i in range(1,times+1):
        closest,centroids,old_centroids,numbers = runKmeans(data,k,1,centroids=centroids,isRandom=False)
        draw_Kmeans(x,closest=closest,centroids=old_centroids,title='Iteration number:'+str(i))
showFigure1(x,3,6)
```

K-均值的一个问题在于，它有可能会停留在一个局部最小值处，而这取决于初始化的情况。
为了解决这个问题，我们通常需要多次运行K-均值算法，每一次都重新进行随机初始化，最后再比较多次运行K-均值的结果，选择代价函数最小的结果。

所以多次运行下面的showFigure1Randomly代码，很有可能会出现不同的结果，有的结果很好有的结果很一般。结果一般的就说明是停留在局部最小值了。

```python
def showFigure1Randomly(data,k,times):
    centroids = None
    for i in range(1,times+1):
        closest,centroids,old_centroids,numbers = runKmeans(data,k,1,centroids=centroids,isRandom=True)
        draw_Kmeans(x,closest=closest,centroids=old_centroids,title='Iteration number:'+str(i))
showFigure1Randomly(x,3,6)
```

## 1.4 Image compression with K-means

用 K-means 压缩图片的思路其实是压缩每个像素，原来每个像素有2的24次方种可能(RGB每个8位)，现在只要16种可能。而具体的做法就是将 RGB 三维空间划分成16个 cluster，然后判断每个像素点属于哪个 cluster就可以了。

[读取的图像矩阵出现小数？使用matplotlib.image.imread()和cv2.imread()读取png图像的差异](https://blog.csdn.net/gyyu32g/article/details/104934667) bird_small.mat中的A和 用 cv2 读取之后转成RGB格式的结果是一样的

```python
# plt.imshow(plt.imread('data/bird_small.png'))
def draw_Figure2(pngPath):
    im = cv2.imread(pngPath,1) # load image as bgr
    im2 = im[:,:,::-1] 	# transform image to rgb
    plt.title('Figure 2:  The original 128x128 image.',y=-0.2)
    plt.imshow(im2)
    plt.show()
    return im2

pngPath = 'data/bird_small.png'
draw_Figure2(pngPath)

fpath = 'data/bird_small.mat'
A = scio.loadmat(fpath)['A']
A.shape
```

## 1.4.1 K-means on pixels

```python
A = A.reshape(-1,3)
A.shape
```

```python
closest,centroids,old_centroids,numbers = runKmeans(A,16,10,isRandom=True)
closest.shape,centroids.shape,old_centroids.shape,numbers
```

这里就已经实现了图片压缩，原来图片占用的字节是 128×128×24 = 393,216 bits.(总共有128×128个像素，每个像素有RGB值，每个值八位)，现在则是 16×24 + 128×128×4 = 65,920 bits. (128×128个像素，每个像素占4位(因为2的4次方是16，每个像素存储的是压缩后16种颜色的下标)，然后16×24 就是这16种颜色的RGB值。)

```python
# 将像素再转回2D
pixels = closest.reshape(128,128)
colors = centroids.copy()
```

```python
# 只能想到for双重循环转。。。 后续看参考答案的时候在学习简单方法吧。
def showCompressedImage(pixels,colors):
    m,n = pixels.shape
    colors = colors / 255 #归一化 plt.imshow() 的参数要不然是[0,1)归一化之后的float（默认matplotlib读取png之后就已经归一化了），要不然是[0,255)原来的int
    baseColorType = colors.shape[1]
    pic = np.zeros(shape=(m,n,baseColorType))
    for i in range(m):
        for j in range(n):
            pic[i][j] = colors[pixels[i][j]]
    plt.title('Figure 3:  Original and reconstructed image (when using K-means to com-press the image).',y=-0.2)
    plt.imshow(pic)
    plt.show()
    
draw_Figure2(pngPath)
showCompressedImage(pixels,colors)
```

## 2.2 Implementing PCA

```python
def featureNormalize(data):
    normalizeData = data.copy()
    normalizeData = (normalizeData - normalizeData.mean(axis=0))/normalizeData.std(axis=0)
    return normalizeData,data.mean(axis=0),data.std(axis=0)
normalizeData,mu,std = featureNormalize(X)
draw_Figure4(normalizeData)
mu,std
```

Σ =$\frac{1}{m}X^TX$ where *X* is the data matrix with examples in rows, and *m* is the number ofexamples.  Note that Σ is a *n×n* matrix and not the summation operator

关于协方差矩阵以及PCA：
+ [浅谈协方差矩阵 ](http://pinkyjie.com/2010/08/31/covariance/)
+ [再谈协方差矩阵之主成分分析 ](http://pinkyjie.com/2011/02/24/covariance-pca/)

```python
def computeCovarianceMatrix(data):
    m = data.shape[0]
    return (data.T.dot(data))/m
# PCA functions ---------------------------------------
covMatrix = computeCovarianceMatrix(normalizeData)
covMatrix
```

关于 svd：
+ [奇异值分解（SVD）](https://zhuanlan.zhihu.com/p/29846048)

(os:研一公共课的矩阵分析其实是有 svd 的，无奈自己当初没好好学。。。另外文章中SVD在PCA中应用时提到特征值和特征向量的意义，这部分本科也欠缺啊。)

```python
U,S,V = np.linalg.svd(covMatrix)
U.shape,S.shape,V.shape
```

```python
U,S
```

```python
def draw_Figure5(data,mu,S,U):
    m = U.shape[0]
    plt.figure(figsize=(6,6))
    
    # Draw the eigenvectors centered at mean of data. These lines show the
    # directions of maximum variations in the dataset.
    # 这是作业中的原话，但是我也没理解了为啥是这两条直线。
    plt.plot((mu[0],mu[0]+1.5*S[0]*U[0,0]),(mu[1],mu[1]+1.5*S[0]*U[0,1]))
    plt.plot((mu[0],mu[0]+1.5*S[1]*U[1,0]),(mu[1],mu[1]+1.5*S[1]*U[1,1]))
    
    plt.scatter(data[:,0],data[:,1],c='white',edgecolors='blue')
    plt.title('Figure 5:  Computed eigenvectors of the dataset',y=-0.2)
    plt.show()
draw_Figure5(X,mu,S,U)
U[0]
```

```python
# 每一分步都完成了，最后是总的pca方法
def pca(data):
    normalizeData,mu,std = featureNormalize(data)
    covMatrix = computeCovarianceMatrix(normalizeData)
    U,S,V =np.linalg.svd(covMatrix)
    return normalizeData,mu,std,U,S,V
normalizeData,mu,std,U,S,V = pca(X)
U
```

## 2.3 Dimensionality Reduction with PCA

### 2.3.1 Projecting the data onto the principal components

```python
def projectData(U,k):
    return U[:,:k]
U_reduce = projectData(U,1)
U_reduce
z = normalizeData[0].dot(U_reduce)
z
```

### 2.3.2 Reconstructing an approximation of the data

```python
def recoverData(z,U_reduce):
    return z.dot(U_reduce.T)
recoverData(z,U_reduce),normalizeData[0]
```

可以看到压缩还原后的数据和元数据的差别还是比较大的，这也很正常，2D -> 1D 都是低维度丢失多了很正常。

### 2.3.3 Visualizing the projections

```python
def draw_Figure6(data,U_reduce):
    plt.figure(figsize=(7,7))
    projectionsData = recoverData(normalizeData.dot(U_reduce),U_reduce)
    plt.scatter(data[:,0],data[:,1],c='white',edgecolors='blue')
    plt.scatter(projectionsData[:,0],projectionsData[:,1],c='white',edgecolors='red')
    m = data.shape[0]
    for i in range(m):
        plt.plot((data[i,0],projectionsData[i,0]),(data[i,1],projectionsData[i,1]),c='black',linestyle='--')
    plt.title('Figure 6: The normalized and projected data after PCA.',y=-0.2)
    plt.show()
draw_Figure6(normalizeData,U_reduce)
```
